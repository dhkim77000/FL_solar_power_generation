```python
#1Î≤à. ÎãπÏßÑ-ÏÑúÎ≤ÑÍ∞Ñ duetÏó∞Í≤∞
import syft as sy
duet = sy.launch_duet(loopback=True)
```

```python
#7Î≤à
import torch
import random
import torch.nn as nn
import torch.nn.functional as F
import pandas as pd
import numpy as np
```

```python
#6Î≤à. ÏÇ¨Ïö©Ìï† Îç∞Ïù¥ÌÑ∞(Î∞úÏ†ÑÎüâ)ÏùÑ Î∂àÎü¨ÏòµÎãàÎã§.
train_out_data_path = r'energy.csv'
train_out_data = pd.read_csv(train_out_data_path)

new_train_out_data=train_out_data.interpolate(method='values')

data_TARGET = np.reshape(np.array(new_train_out_data['dangjin'].values)+np.array(new_train_out_data['dangjin_floating'].values)+np.array(new_train_out_data['dangjin_warehouse'].values),(1068,24))/2700

train_len=848
valid_len=110

#Îç∞Ïù¥ÌÑ∞Î•º Ìä∏Î†àÏù¥Îãù,Î∞úÎ¶¨Îç∞Ïù¥ÏÖò,ÌÖåÏä§Ìä∏ ÏÖãÏúºÎ°ú ÎÇòÎàïÎãàÎã§.
train_output=torch.FloatTensor(data_TARGET[:train_len,:])
train_output = train_output.tag("Dangjin Train Data")

valid_output=torch.FloatTensor(data_TARGET[train_len:train_len+valid_len,:])
valid_output = valid_output.tag("Dangjin Valid Data")

test_output=torch.FloatTensor(data_TARGET[train_len+valid_len:,:])
test_output = test_output.tag('Dangjin Test Data')

#ÏÑ∏Í∞úÏùò Îç∞Ïù¥ÌÑ∞ ÏÖãÏùÑ pysyftÏÉÅÏóê Ïò¨Î¶ΩÎãàÎã§.
train_output.send(duet,pointable=True)
valid_output.send(duet,pointable=True)
test_output.send(duet,pointable=True)

#ÏÑúÎ≤ÑÏóêÏÑú Îç∞Ïù¥ÌÑ∞Ïùò Ï†ëÏÜçÏùÑ ÌóàÍ∞ÄÌï¥Ï§çÎãàÎã§.
duet.requests.add_handler(
    action="accept",
    print_local=True,  # print the result in your notebook
)
```

```python
#3번. 울산-서버간 duet연결
import syft as sy
duet = sy.launch_duet(loopback=True)
```

```python
#7번
import torch
import random
import torch.nn as nn
import torch.nn.functional as F
import pandas as pd
import numpy as np
from torch.utils.data import Dataset 
```

```python
#8번. 사용할 데이터(발전량)을 불러옵니다.
train_out_data_path = r'energy.csv'
train_out_data = pd.read_csv(train_out_data_path)

new_train_out_data=train_out_data.interpolate(method='values')

data_TARGET = np.reshape(np.array(new_train_out_data['ulsan'].values),(1068,24))/500

train_len=848
valid_len=110

#데이터를 트레이닝,발리데이션,테스트 셋으로 나눕니다.
train_output=torch.FloatTensor(data_TARGET[:train_len,:])
train_output = train_output.tag("Ulsan Train Data")

valid_output=torch.FloatTensor(data_TARGET[train_len:train_len+valid_len,:])
valid_output = valid_output.tag('Ulsan Valid Data')

test_output= torch.FloatTensor(data_TARGET[train_len+valid_len:,:])
test_output = test_output.tag("Ulsan Test Data")

#세개의 데이터 셋을 pysyft상에 올립니다.
train_output.send(duet,pointable=True)
valid_output.send(duet,pointable=True)
test_output.send(duet,pointable=True)

#서버에서 데이터의 접속을 허가해줍니다.
duet.requests.add_handler(
    action="accept",
    print_local=True,  # print the result in your notebook
)
```
